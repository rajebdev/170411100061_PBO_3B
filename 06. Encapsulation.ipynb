Encapsulation adalah proses penyembunyian property atau method pada suatu kelas agar tidak dapat diakses dari luar kelas.  Proses encapsulation ditujukan agar proses yang tidk harus diketahui oleh pengguna bisa terjaga secara aman. seperti ketika mobil menyala, mesin akan otomatis memulati startEngine sehingga mobil bisa menyala.

<img src="https://1.bp.blogspot.com/-6j04YIx5-qM/W8wPAVh39DI/AAAAAAAACHs/AY3Ti3FAWCUJLHH4LX3u5_GlmCdl3roFwCLcBGAs/s1600/7.1.png"/>


Beberapa istilah dalam encapsulation :
1. Private method adalah sebuah method yang disembunyikan agar tidak bisa diakses dari luar kelas secara langsung. 
2. Private Attribut/Property adalah sebuah Atribut/property yang disembunyikan sehingga nilainya tidak diketahui dari luar class secara langsung.
3. Public Method  dan Public Atribut/Property dapat diakses dari mana saja baik dari dalam maupun luar kelas
4. Private Method dan Private Atribut Hanya dapat diakses didalam kelas itu sendiri.



Untuk mengakses property/Atribut dan method yang terencapsulation ada 3 cara yaitu :
1. Dengan Membuat property/method baru dengan isi/nilai di dalamnya yaitu method/property yang terencapsulation

2. Dengan Menggunakan Method baru Tambahan
    - Membuat sebuah method bernama get... dengan isi return nilai dari property yang ingin agar bisa di akses dari luar
    
3. Dengan Menambahkan _NamaKelas Di depan penulisan nama method yang terencapsulation
    - Melakukan pemanggilan method dari kelas objek seperti biasa kemudian ditambahn _NamaKelas didepannya. (eg. redcar._Car__speed untuk property atau redcar._Car__startEngine() untuk method)

```python
#Contoh Program Encapsulation Python

class Car:
    #__speed = 0
    #__name = ""
    
    def __init__(self, velocity, name):
        self.__speed = velocity
        self.speed = self.__speed
        self.__name = name
        #self.__startEngine()
        
    def showData(self):
        print('name %s, speed %s'%(self.__speed, self.__name))
    
    def getSpeed(self):
        return self.__speed
    
    def __startEngine(self):
        print('Engine Started')

redcar = Car(300,'ferari')

redcar.showData()
print(redcar.getSpeed())
print(redcar._Car__speed)
```

```python
#Contoh Modifikasi Kelas Biasa Menjadi Encapsulation

class Makanan:
    def __init__(self, namaBrg, harga, diskon, exDate):
        
        #private property
        self.__kode = 'CB0001'
        self.__deskripsi ="Makanan"
        
        #public property
        self.namaBrg = namaBrg
        self.harga = harga
        self.diskon = diskon
        self.exDate = exDate
        self.__showDeskripsi()
        
        #private to public
        self.kode = self.__kode
        self.deskripsi = self.__deskripsi
        
    def TampilSemua(self):
        print('\nKode\t\t: %s\nKatergory\t: %s\n' \
              'Nama Barang\t: %s\nHarga Barang\t: Rp %s,-\nDiskon Barang\t: %s\n' \
              'Expire Date\t: %s\n' \
              %(self.__kode,self.__deskripsi,self.namaBrg,self.harga, self.diskon, self.exDate))
    
    def setKode(self, kode):
        print('Anda Mengubah Kode dari', self.__kode,'menjadi',kode)
        self.__kode = kode
    
    def getKode(self):
        return self.__kode

    def __showDeskripsi(self):
        print('Anda Memilih',self.__deskripsi,'\n')
    
    def Deskripsi(self):
        self.__showDeskripsi()
    
roti = Makanan('Sari Roti', 4999, '10%', '17102018')
roti.TampilSemua()


print('Kode Awal :',roti.getKode())

#Perubahan Data Encapsulation dengan dibantu method tambahan set.
roti.setKode('CB0003')

#pemanggilan Encapsulation dengan metode method tambahan yaitu get
print('Kode Setelah diubah :',roti.getKode())

#pemanggilan Encapsulation dengan metode penambahan _NamaKelas didepan nama method atau property
roti._Makanan__showDeskripsi()

roti.TampilSemua()

print(roti.kode,roti.deskripsi)

        
```
