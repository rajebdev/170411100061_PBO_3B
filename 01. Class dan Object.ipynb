C. Istilah – Istilah Dalam OOP

Kita harus mengenal beberapa istilah dalam pemrograman berbasis object :

- Kelas – Kelas adalah pondasi dari sebuah pembentukan object.
- Variabel Kelas – Variabel kelas adalah sebuah penyimpanan yang dapat diakses dari dalam class atau luar class, walaupun ada juga variabel private yg hanya bisa diakses pada class tersebut.
- Variabel instansiasi – Variabel instansiasi adalah sebuah variabel yang di buat di dalam sebuah method tanpa menggunakan self. sehingga hanya dapat di akses di method tersebut.
- Instance – Instance adalah nama lain dari objek.
- Instansiasi – Instansiasi adalah proses pembuatan sebuah object atau instance.
- Metode – Metode adalah sebuah fungsi yang dibuat di dalam kelas sehingga dapat dipanggil berdasarkan fungsinya.
- Objek – Objek adalah hasil dari pembentukan sebuah class. Dimana dari satu kelas bisa di buat menjadi banyak sekali objek berbeda.

```python
print('Nomer 1 : Class Kubus')

class Kubus():
    def __init__(self,s):
        self.sisi = s
    def tampilkansisi(self):
        print("Sisi = ", self.sisi)
    def luasSisi(self):
        print("Luas = ", self.sisi**2)
    
    def volume(self):
        print("Volume = ", self.sisi ** 3)
        
    def luasPermukaan(self):
        print("Luas Permukaan = ", (self.sisi**2)*6 )
        
kubusBesar = Kubus(100)
kubusBesar.tampilkansisi();
kubusBesar.luasSisi();
kubusBesar.volume()
kubusBesar.luasPermukaan()



print('\nNomer 2 : Class Mahasiswa')

class Mahasiswa():
    def __init__(self, nim, nama, ipk):
        self.nim = nim;
        self.nama = nama;
        self.ipk = ipk;
        
    def tampilkan(self):
        print('NIM  : %s \nNama : %s \nIPK  : %s' % (self.nim,self.nama, self.ipk))
    
    def predikat(self):
        ipkMHS = self.ipk;
        if ipkMHS >= 3.5 :
            print("Predikat : Cumlaude");
        elif 3 < ipkMHS < 3.5 :
            print("Predikat : Sangat Memuaskan");
        elif 2.5 < ipkMHS <= 3 :
            print("Predikat : Memuaskan");
        else:
            print("Predikat : Cukup");

            
Hilal = Mahasiswa(17053,'Hilaludin',3.6)
Hilal.tampilkan();
Hilal.predikat()

```

#### Perbedaan Class & Object ####






- Class adalah sebuah code template untuk membuat sebuah objek.<br>
- Objek  adalah hasil dari bentukan class.


```python
print('Nomer 3 : Class Nasabah dan Rekening')

class Nasabah():
    def __init__(self, nomor=[], nama=[], norek=[]):
        self.nomor = nomor;
        self.nama = nama;
        self.norek = norek;
        
    def tampilkan(self):
        print('\n\nNomor  : %s \nNama : %s \nNo. Rekening  : %s' % (self.nomor,self.nama, self.norek))
    
        
class Rekening():
    def __init__(self, norek=[], saldo=[]):
        self.norek = norek;
        self.saldo = saldo;
        
    def ambilUang(self, a_nom):
        self.saldo -= a_nom;
        print("\nJumlah Penarikan : Rp.", a_nom)
        print("Saldo Sekarang : Rp.", self.saldo)
        
    def setorUang(self, s_nom):
        self.saldo += s_nom;
        print("\nJumlah Setoran : Rp.", s_nom)
        print("Saldo Sekarang : Rp.", self.saldo)
        
    def cekSaldo(self):
        print("\nSisa Saldo ",self.norek," = Rp.",self.saldo);
        
#Nasabah        
Hendrik = Nasabah(2,'Hendrik Wicaksono', 170411100057)
Rohman = Nasabah(3, 'Rohman Fatoni', 170411100095)
#Rekening
hen_rek = Rekening(170411100057, 3000000)
rom_rek = Rekening(170411100095, 1000000)

#Simulasi hendrik
Hendrik.tampilkan()
hen_rek.ambilUang(200000)
hen_rek.setorUang(150000)
hen_rek.cekSaldo()

#Simulasi Rohhman
Rohman.tampilkan()
hen_rek.ambilUang(350000)
rom_rek.setorUang(200000)
rom_rek.cekSaldo()
```
